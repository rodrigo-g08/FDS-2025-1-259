```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seleccionamos solo las columnas numéricas
columnas_numericas = df.select_dtypes(include=['number']).columns
```

```python
# Tamaño general del gráfico
plt.figure(figsize=(15, 5 * len(columnas_numericas)))

# Iterar por cada columna numérica y graficar
for i, columna in enumerate(columnas_numericas, 1):
    plt.subplot(len(columnas_numericas), 1, i)
    sns.boxplot(data=df, x=columna, color="lightblue")
    plt.title(f"Boxplot de {columna}")
    plt.tight_layout()

# Filtramos solo columnas numéricas
columnas_numericas = df.select_dtypes(include=['number']).columns

# Diccionario para guardar el conteo de outliers
conteo_outliers = {}

# Recorremos cada columna numérica
for columna in columnas_numericas:
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    # Condición de outlier
    outliers = df[(df[columna] < Q1 - 1.5 * IQR) | (df[columna] > Q3 + 1.5 * IQR)]
    conteo_outliers[columna] = len(outliers)

# Mostrar resultados ordenados
outliers_df = pd.DataFrame.from_dict(conteo_outliers, orient='index', columns=['Cantidad de Atipicos'])
outliers_df = outliers_df.sort_values(by='Cantidad de Atipicos', ascending=False)

print(outliers_df)
```

WINSORIZAR Y COMPLETAR LOS VALORES VACIOS

```python
#Para la columna descripcion
df['description'] = df['description'].fillna("Sin descripción")
```

```python
columnas_atipicas = ['dislikes', 'views', 'comment_count', 'likes', 'lon']

def winsorizar_columna(columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    # Reemplaza valores extremos con los límites
    return df[columna].clip(lower=limite_inferior, upper=limite_superior)

# Creamos nuevas columnas winsorizadas
for columna in columnas_atipicas:
    
    df[columna] = winsorizar_columna(columna)
```

```python
df.to_csv("GBvideos_cc50_202101_limpio.csv", index=False, encoding='utf-8')
```
